# DATA CLEANING & DATA MANIPULATION

```python
Data.isna().sum()
```

```python
Data = Data.dropna()
Data
```

```python
Data['Order Date'].head()
```

```python
Data['Order Date'] = pd.to_datetime(Data['Order Date'],format='%m/%d/%y %H:%M',errors = 'coerce')
Data
```

```python
Data['Street Number'] = Data['Purchase Address'].str.split(' ', n=1).str[0]
Data['Address'] = Data['Purchase Address'].str.split(' ', n=1).str[1]
Data.drop('Purchase Address',axis=1 , inplace=True)
Data
```

```python
split_address = Data['Address'].str.split(',', expand=True)
Data['Street Name'] = split_address[0]
Data['City'] = split_address[1]
Data['State'] = split_address[2]
Data
```

```python
Data.drop('Address',axis=1,inplace=True)
```

```python
Data
```

```python
State = Data['State'].str.split(' ', expand=True)
Data.drop('State',axis=1,inplace=True)
```

```python
Data['State'] = State[1]
Data['Postal code'] = State[2]
Data
```

```python
Data['Quantity Ordered'] = Data['Quantity Ordered'].str.replace('[^\d.]', '', regex=True)
Data['Quantity Ordered'] = pd.to_numeric(Data['Quantity Ordered'], errors='coerce')
Data = Data.dropna(subset=['Quantity Ordered'])
Data['Price Each'] = Data['Price Each'].str.replace(' ', '', regex=True).astype(float)
Data.info()
```

```python
Data['Total Price'] = Data['Quantity Ordered'] * Data['Price Each']
Data
```

Load

```python
Data.to_csv('FullData.csv')
```
